```python
import tkinter as tk
from tkinter import ttk
import pandas as pd

class App(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Pandas Table")
        self.df = pd.read_csv("products.csv")
        # Create Dropdown Menus
        self.search_options = ttk.Combobox(self, width=20)
        self.search_options["values"] =  list( self.df['product_type'].unique())
        self.search_options.pack(pady=10)


         # Create Search button
        self.search_button = ttk.Button(self, text="Search", command=self.search_data)
        self.search_button.pack(pady=10)
        
        # Create Treeview
        self.treeview = ttk.Treeview(self)
        self.treeview.pack(side="left", fill="both", expand=True)
        self.scrollbar = ttk.Scrollbar(self, orient="vertical", command=self.treeview.yview)
        self.scrollbar.pack(side="right", fill="y")
        self.treeview.configure(yscrollcommand=self.scrollbar.set)

        # Load data
        self.df = pd.read_csv("products.csv")
        self.treeview["columns"] = list(self.df.columns)
        for column in self.treeview["columns"]:
            self.treeview.heading(column, text=column)
        self.df_rows = self.df.to_numpy().tolist()
        self.display_all_data()

        # Create Reload button
        self.reload_button = ttk.Button(self, text="Reload", command=self.reload_data)
        self.reload_button.pack(pady=10)

    def search_data(self):
        search_term = self.search_options.get() 
        filtered_data =  self.df[self.df['product_type'] ==str(search_term)].to_numpy().tolist()
        self.treeview.delete(*self.treeview.get_children())
        for row in filtered_data:
            self.treeview.insert("", "end", values=row)

    def display_all_data(self):
        self.treeview.delete(*self.treeview.get_children())
        for row in self.df_rows:
            self.treeview.insert("", "end", values=row)

    def reload_data(self):
        self.df = pd.read_csv("products.csv")
        self.treeview["columns"] = list(self.df.columns)
        for column in self.treeview["columns"]:
            self.treeview.heading(column, text=column)
        self.df_rows = self.df.to_numpy().tolist()
        self.display_all_data()

app = App()
app.mainloop()
```

```python
import tkinter as tk
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Tạo cửa sổ chính
window = tk.Tk()
window.title("Thống kê báo cáo sản phẩm")

# Tạo hộp văn bản để nhập dữ liệu
entry_data = tk.Entry(window, width=100)
entry_data.pack()

# Tạo nút để bắt đầu phân tích
button_start = tk.Button(window, text="Bắt đầu", command=lambda: start_analysis())
button_start.pack()

# Tạo hộp văn bản để hiển thị kết quả
text_result = tk.Text(window, width=100, height=20)
text_result.pack()

# Hàm bắt đầu phân tích
def start_analysis():
    # Nhận dữ liệu từ hộp văn bản
    data = pd.read_csv("products.csv").to_numpy()
    print(data)

    # Chuyển đổi dữ liệu thành mảng NumPy
   
    # Thống kê dữ liệu
    mean = np.mean(data)
    std = np.std(data)

    # Tạo biểu đồ phân bố dữ liệu
    plt.hist(data)
    plt.axvline(mean, color="red")
    plt.axvline(mean + std, color="green")
    plt.axvline(mean - std, color="green")
    plt.show()

    # Hiển thị kết quả
    text_result.delete(1.0, "end")
    text_result.insert(1.0, "Kết quả thống kê: \n\n"
                         "Trung bình: {:.2f}\n"
                         "Độ lệch chuẩn: {:.2f}".format(mean, std))

# Chạy chương trình
window.mainloop()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk

# Tạo dữ liệu
x = np.arange(0, 10)
y = x ** 2

# Tạo các hàm vẽ đồ thị
def plot_line(x, y):
    plt.plot(x, y)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()

def plot_bar(x, y):
    plt.bar(x, y)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()

def plot_pie(x, y):
    plt.pie(y, labels=x)
    plt.show()

# Tạo giao diện người dùng
root = tk.Tk()

# Tạo các nút
line_button = tk.Button(root, text="Biểu đồ đường", command=lambda: plot_line(x, y))
bar_button = tk.Button(root, text="Biểu đồ cột", command=lambda: plot_bar(x, y))
pie_button = tk.Button(root, text="Biểu đồ tròn", command=lambda: plot_pie(x, y))

# Sắp xếp các nút
line_button.pack()
bar_button.pack()
pie_button.pack()

# Khởi động vòng lặp chính
root.mainloop()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk

# Tạo dữ liệu
x = np.arange(0, 10)
y = x ** 2

# Tạo giao diện người dùng
root = tk.Tk()

# Tạo canvas
canvas = tk.Canvas(root, width=500, height=500)

# Vẽ biểu đồ đường
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Thêm canvas vào giao diện người dùng
canvas.pack()

# Cập nhật biểu đồ đường mỗi giây
def update_graph():
    global x, y

    x = x + 1
    y = x ** 2

    # Vẽ biểu đồ đường mới
    plt.cla()
    plt.plot(x, y)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()

    # Đặt lại thời gian chờ
    canvas.after(1000, update_graph)

# Khởi động vòng lặp chính
update_graph()
root.mainloop()
```
